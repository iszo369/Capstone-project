## **Kenya Regional Crop Yield Prediction Using Machine Learning**

### Background (Business Understanding)

Agriculture is one of the leading economic drivers in Kenya. According to the World Bank [^1], the sector contributes **20% - 30% of the total national Gross Domestic Product (GDP)**. This means that agriculture significantly tips the economic balance in terms of value added to the economy through **employment, food security, and poverty reduction**.  

While agriculture as a whole has huge significance for the economy, it is vital to understand that the **crop farming sub-sector contributes approximately 15% to the entire national GDP**[^2] . This highlights the specific impact of crop farming within the broader agricultural sector.

It is the leading subsector in agriculture followed by livestock farming,forestry and finally fishing.  Therefore crop farming is one of the key indirect drivers of the economy. Resent publiction [^3],  crop farming is directly tied to the fast economic growth via agricuture in first quater of 2025. This underpins the fact that high crop yield coinsides with higher economic growth. 

Therefore the possiblities that would appear as a result of predicting crop yields are quite endless, the following are just afew solutions that would be possible by predicting crop yields in Kenya ;

1. Provide farm level intelligence to farmers in terms of crop variety with possible highest yield
2. National food security informed decision making in form of early warning for poor harvest season
3. Help inform economic policies such as when to provide subbsidies to farmers 
4. Assist iin better storage and logistic preparation
5. The model can help power a digital agricultural product such as Mobile advisory app for farmers.
